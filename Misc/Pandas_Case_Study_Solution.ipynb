```python
import pandas as pd
import numpy as np

# For reproducibility
np.random.seed(42)

# Number of synthetic rows
n = 10

# Generate synthetic data
data = {
    "EmployeeID": range(1, n+1),
    "Name": [f"Emp{i}" for i in range(1, n+1)],
    "Age": np.random.randint(22, 60, n),                    # random ages
    "Department": np.random.choice(["HR", "IT", "Finance", "Marketing"], n),
    "JoiningDate": pd.to_datetime("2020-01-01") + pd.to_timedelta(np.random.randint(0, 1500, n), unit="D"),
    "Salary": np.random.randint(30000, 120000, n),          # random salary
    "PerformanceScore": np.round(np.random.uniform(1, 5, n), 2),  # random float
    "IsActive": np.random.choice([True, False], n)          # boolean
}

# Create DataFrame
df = pd.DataFrame(data)

print(df)
```

```python
for col in df.columns:
    print(f"\nColumn: {col}")
    print(df[col].value_counts())
```

```python
df1 = pd.DataFrame({
    "ID": [1, 2, 3],
    "Dept": ["HR", "IT", "IT"],
    "Name": ["Alice", "Bob", "Charlie"]
})

df2 = pd.DataFrame({
    "ID": [1, 2, 3],
    "Dept": ["HR", "IT", "Finance"],
    "Salary": [50000, 60000, 70000]
})
```

```python
merged = pd.merge(df1, df2, how='inner', on=["ID"])
```

```python
print(merged)
```

```python
df1 = pd.DataFrame({"emp_id": [1,2], "Name": ["Alice","Bob"]})
df2 = pd.DataFrame({"employee_id": [1,3], "Salary": [50000,70000]})

merged = pd.merge(df1, df2, left_on="emp_id", right_on="employee_id", how="inner")
print(merged)
```
