````python
import numpy as np # Store Arrays
import matplotlib.pyplot as plt # Ploting Purpose
import matplotlib.patches as patches
from matplotlib_venn import venn2 # Venn Diagram creation
import seaborn as sns# Visual Representation
````

````python
print("\n➕ RULE 2: ADDITION RULE")
print("=" * 40)

print("🎯 THE IDEA:")
print("When you want the probability of A OR B happening, you add their probabilities.")
print("BUT be careful not to double-count if they can happen together!")
print()
print("Formula: P(A OR B) = P(A) + P(B) - P(A AND B)")
print("If A and B can't happen together: P(A OR B) = P(A) + P(B)")

# Example 1: Mutually Exclusive Events
print("\n🎲 EXAMPLE 1: ROLLING A DIE (Mutually Exclusive)")
print("What's the probability of rolling a 1 OR a 2?")
print("These can't happen at the same time, so we just add!")

prob_1 = 1/6
prob_2 = 1/6
prob_1_or_2 = prob_1 + prob_2

print(f"P(Roll 1) = {prob_1:.3f}")
print(f"P(Roll 2) = {prob_2:.3f}")
print(f"P(Roll 1 OR 2) = {prob_1:.3f} + {prob_2:.3f} = {prob_1_or_2:.3f}")
print("This makes sense: 2 favorable outcomes out of 6 total!")

# Example 2: Non-Mutually Exclusive Events
print("\n🃏 EXAMPLE 2: DRAWING CARDS (Can Overlap)")
print("What's the probability of drawing a Heart OR a Face card?")

# In a standard deck: 52 cards, 13 hearts, 12 face cards, 3 face cards that are hearts
prob_heart = 13/52
prob_face = 12/52
prob_heart_and_face = 3/52  # Jack, Queen, King of Hearts
prob_heart_or_face = prob_heart + prob_face - prob_heart_and_face

print(f"P(Heart) = 13/52 = {prob_heart:.3f}")
print(f"P(Face card) = 12/52 = {prob_face:.3f}")
print(f"P(Heart AND Face) = 3/52 = {prob_heart_and_face:.3f}")
print(f"P(Heart OR Face) = {prob_heart:.3f} + {prob_face:.3f} - {prob_heart_and_face:.3f} = {prob_heart_or_face:.3f}")
print("We subtract the overlap to avoid counting the same cards twice!")

# Visualize addition rule with Venn diagrams
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Mutually exclusive
ax1.add_patch(patches.Rectangle((0.2, 0.3), 0.2, 0.4, facecolor='lightblue', alpha=0.7, label='Roll 1'))
ax1.add_patch(patches.Rectangle((0.6, 0.3), 0.2, 0.4, facecolor='lightcoral', alpha=0.7, label='Roll 2'))
ax1.set_xlim(0, 1)
ax1.set_ylim(0, 1)
ax1.set_title('Mutually Exclusive Events\n(No Overlap)')
ax1.text(0.3, 0.5, 'Roll 1', ha='center', va='center', fontweight='bold')
ax1.text(0.7, 0.5, 'Roll 2', ha='center', va='center', fontweight='bold')
ax1.text(0.5, 0.1, 'P(1 OR 2) = P(1) + P(2)', ha='center', fontsize=12)

# Non-mutually exclusive using venn diagram
ax2.text(0.5, 0.9, 'Hearts OR Face Cards', ha='center', fontsize=14, fontweight='bold')
venn = venn2(subsets=(10, 9, 3), set_labels=('Hearts', 'Face Cards'), ax=ax2)
venn.get_patch_by_id('10').set_color('lightblue')
venn.get_patch_by_id('01').set_color('lightcoral')
venn.get_patch_by_id('11').set_color('purple')
ax2.text(0.5, 0.1, 'P(H OR F) = P(H) + P(F) - P(H AND F)', ha='center', fontsize=10)

plt.tight_layout()
plt.show()
````

````python
print("\n🔄 RULE 1: COMPLEMENT RULE")
print("=" * 40)

print("🎯 THE IDEA:")
print("If something has a certain probability, the probability of it NOT happening")
print("is just 1 minus that probability!")
print()
print("Formula: P(NOT A) = 1 - P(A)")
print("Or: P(A) + P(NOT A) = 1")

# Example 1: Weather
print("\n🌤️  EXAMPLE 1: WEATHER")
print("If there's a 30% chance of rain...")
prob_rain = 0.3
prob_no_rain = 1 - prob_rain

print(f"P(Rain) = {prob_rain}")
print(f"P(No Rain) = 1 - {prob_rain} = {prob_no_rain}")
print(f"Check: {prob_rain} + {prob_no_rain} = {prob_rain + prob_no_rain} ✓")

# Example 2: Dice
print("\n🎲 EXAMPLE 2: ROLLING A DIE")
print("What's the probability of NOT rolling a 6?")
prob_six = 1/6
prob_not_six = 1 - prob_six

print(f"P(Rolling 6) = 1/6 = {prob_six:.3f}")
print(f"P(NOT Rolling 6) = 1 - 1/6 = {prob_not_six:.3f}")
print(f"This makes sense: there are 5 ways to NOT roll a 6 out of 6 total!")

# Visualize complement rule
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Weather example
ax1.pie([prob_rain, prob_no_rain], labels=['Rain', 'No Rain'],
        colors=['lightblue', 'orange'], autopct='%1.1f%%', startangle=90)
ax1.set_title('Weather Probability\n(Complement Rule)')

# Dice example
dice_probs = [prob_six] + [prob_not_six/5]*5
labels = ['Roll 6'] + [f'Roll {i}' for i in range(1,6)]
colors = ['red'] + ['lightgreen']*5

ax2.pie(dice_probs, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
ax2.set_title('Dice Probability\n(6 vs NOT 6)')

plt.tight_layout()
plt.show()
````

````python
# =============================================================================
# EXAMPLE 2: Medical Testing (Classic Example)
# =============================================================================
print("\n🏥 EXAMPLE 2: MEDICAL TESTING")
print("=" * 40)

print("This is the classic example that surprises everyone!")
print()
print("Scenario:")
print("• A disease affects 1% of the population")
print("• A test is 99% accurate (detects disease when present)")
print("• The test has 1% false positive rate (says disease when healthy)")
print()
print("Question: If you test positive, what's the probability you have the disease?")
print()
print("Most people guess 99%, but let's see...")

# Medical testing probabilities
disease_rate = 0.01
test_sensitivity = 0.99  # True positive rate
false_positive_rate = 0.01

# Calculate using Bayes
prob_positive = test_sensitivity * disease_rate + false_positive_rate * (1 - disease_rate)
prob_disease_given_positive = (test_sensitivity * disease_rate) / prob_positive

print(f"📊 Given:")
print(f"• P(Disease) = {disease_rate} (1% of population)")
print(f"• P(Positive test | Disease) = {test_sensitivity} (99% accurate)")
print(f"• P(Positive test | Healthy) = {false_positive_rate} (1% false positive)")
print()
print(f"🧮 Using Bayes:")
print(f"• P(Positive test) = {test_sensitivity} × {disease_rate} + {false_positive_rate} × {1-disease_rate}")
print(f"                   = {prob_positive:.4f}")
print()
print(f"• P(Disease | Positive) = {test_sensitivity} × {disease_rate} / {prob_positive:.4f}")
print(f"                        = {prob_disease_given_positive:.4f}")
print()
print(f"🎯 SHOCKING RESULT: Only {prob_disease_given_positive:.1%} chance you have the disease!")
print("Even with a 99% accurate test!")

# Visualize medical example
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5))

# Population breakdown
population = 10000
diseased = int(population * disease_rate)
healthy = population - diseased

true_positives = int(diseased * test_sensitivity)
false_negatives = diseased - true_positives
false_positives = int(healthy * false_positive_rate)
true_negatives = healthy - false_positives

# Population visualization
ax1.bar(['Diseased', 'Healthy'], [diseased, healthy], color=['red', 'green'], alpha=0.7)
ax1.set_title(f'Population of {population:,}')
ax1.set_ylabel('Number of People')
for i, v in enumerate([diseased, healthy]):
    ax1.text(i, v + 200, str(v), ha='center', fontweight='bold')

# Test results
categories = ['True +', 'False +', 'True -', 'False -']
values = [true_positives, false_positives, true_negatives, false_negatives]
colors = ['darkgreen', 'red', 'lightgreen', 'orange']

ax2.bar(categories, values, color=colors, alpha=0.7)
ax2.set_title('Test Results')
ax2.set_ylabel('Number of People')
for i, v in enumerate(values):
    ax2.text(i, v + 100, str(v), ha='center', fontweight='bold')

# Focus on positive tests
positive_tests = true_positives + false_positives
ax3.pie([true_positives, false_positives],
        labels=[f'Actually Diseased\n{true_positives}', f'Actually Healthy\n{false_positives}'],
        colors=['red', 'lightcoral'], autopct='%1.1f%%', startangle=90)
ax3.set_title(f'Among {positive_tests} Positive Tests\nOnly {prob_disease_given_positive:.1%} are real!')

plt.tight_layout()
plt.show()

print(f"\n💡 WHY THIS HAPPENS:")
print(f"• The disease is rare (only {diseased} out of {population:,} people)")
print(f"• Even with 1% false positives, that's {false_positives} healthy people testing positive")
print(f"• Only {true_positives} diseased people test positive")
print(f"• So most positive tests ({false_positives} out of {positive_tests}) are false alarms!")
````

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew
import seaborn as sns
from collections import Counter

print("📊 SKEWNESS ANALYSIS - Understanding Your Dataset")
print("=" * 60)

# The dataset
dataset = [88, 85, 82, 97, 67, 77, 74, 86, 81, 95, 77, 88, 85, 76, 81]

print("\n📋 YOUR DATASET:")
print("-" * 20)
print("Raw data:", dataset)
print("Sorted data:", sorted(dataset))
print()

# Basic statistics
n = len(dataset)
mean_val = np.mean(dataset)
median_val = np.median(dataset)
std_val = np.std(dataset)
min_val = min(dataset)
max_val = max(dataset)
range_val = max_val - min_val

print("📈 BASIC STATISTICS:")
print(f"• Count: {n}")
print(f"• Mean: {mean_val:.2f}")
print(f"• Median: {median_val:.2f}")
print(f"• Standard deviation: {std_val:.2f}")
print(f"• Minimum: {min_val}")
print(f"• Maximum: {max_val}")
print(f"• Range: {range_val}")

# Calculate skewness
skewness = skew(dataset, axis=0, bias=True)
print(f"\n🎯 SKEWNESS: {skewness:.4f}")

# =============================================================================
# UNDERSTANDING SKEWNESS
# =============================================================================
print("\n🤔 WHAT IS SKEWNESS?")
print("-" * 25)

print("Skewness measures the asymmetry of a distribution:")
print()
print("📏 SKEWNESS VALUES:")
print("• Skewness = 0: Perfectly symmetric (like a normal bell curve)")
print("• Skewness > 0: Positively skewed (right-tailed)")
print("• Skewness < 0: Negatively skewed (left-tailed)")
print()
print("📊 RULE OF THUMB:")
print("• -0.5 to 0.5: Approximately symmetric")
print("• 0.5 to 1.0: Moderately positively skewed")
print("• > 1.0: Highly positively skewed")
print("• -1.0 to -0.5: Moderately negatively skewed")
print("• < -1.0: Highly negatively skewed")

# =============================================================================
# WHY YOUR DATASET IS POSITIVELY SKEWED
# =============================================================================
print(f"\n🔍 WHY YOUR DATASET IS POSITIVELY SKEWED ({skewness:.4f})")
print("-" * 50)

print("Let's examine the data more closely:")
print()

# Sort and analyze
sorted_data = sorted(dataset)
print("📊 SORTED DATA ANALYSIS:")
print(f"Sorted: {sorted_data}")
print()

# Find quartiles
q1 = np.percentile(dataset, 25)
q3 = np.percentile(dataset, 75)
iqr = q3 - q1

print(f"• Q1 (25th percentile): {q1}")
print(f"• Q2 (50th percentile/median): {median_val}")
print(f"• Q3 (75th percentile): {q3}")
print(f"• IQR (Q3 - Q1): {iqr}")
print()

# Analyze the distribution
print("🎯 KEY OBSERVATIONS:")
print()

# Count values in different ranges
lower_half = [x for x in dataset if x < median_val]
upper_half = [x for x in dataset if x > median_val]
at_median = [x for x in dataset if x == median_val]

print(f"• Values below median ({median_val}): {len(lower_half)} values")
print(f"  → {lower_half}")
print(f"• Values at median: {len(at_median)} values")
print(f"• Values above median: {len(upper_half)} values")
print(f"  → {upper_half}")
print()

# Distance analysis
print("📏 DISTANCE FROM MEAN ANALYSIS:")
distances = [(x, x - mean_val, abs(x - mean_val)) for x in dataset]
distances.sort(key=lambda x: x[1])  # Sort by distance from mean

print("Value | Distance from Mean | Absolute Distance")
print("-" * 45)
for val, dist, abs_dist in distances:
    direction = "above" if dist > 0 else "below" if dist < 0 else "at"
    print(f"{val:5d} | {dist:+8.2f}        | {abs_dist:8.2f}")

# =============================================================================
# VISUAL ANALYSIS
# =============================================================================
print(f"\n📊 VISUAL ANALYSIS")
print("-" * 20)

# Create comprehensive visualization
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 12))

# 1. Histogram
ax1.hist(dataset, bins=8, alpha=0.7, color='skyblue', edgecolor='black')
ax1.axvline(mean_val, color='red', linestyle='--', linewidth=2, label=f'Mean: {mean_val:.2f}')
ax1.axvline(median_val, color='green', linestyle='--', linewidth=2, label=f'Median: {median_val:.2f}')
ax1.set_xlabel('Values')
ax1.set_ylabel('Frequency')
ax1.set_title('Histogram showing Right Skew')
ax1.legend()
ax1.grid(True, alpha=0.3)

# Add skew annotation
ax1.text(0.05, 0.95, f'Skewness: {skewness:.4f}\n(Positive = Right Skew)',
         transform=ax1.transAxes, fontsize=10, verticalalignment='top',
         bbox=dict(boxstyle='round', facecolor='yellow', alpha=0.7))

# 2. Box plot
box_data = ax2.boxplot(dataset, vert=True, patch_artist=True)
box_data['boxes'][0].set_facecolor('lightblue')
ax2.set_ylabel('Values')
ax2.set_title('Box Plot showing Outliers')
ax2.grid(True, alpha=0.3)

# Add statistics to box plot
ax2.text(1.2, q1, f'Q1: {q1}', va='center')
ax2.text(1.2, median_val, f'Median: {median_val}', va='center')
ax2.text(1.2, q3, f'Q3: {q3}', va='center')

# 3. Dot plot showing individual values
ax3.scatter(dataset, [1] * len(dataset), alpha=0.7, s=100, color='blue')
ax3.axvline(mean_val, color='red', linestyle='--', linewidth=2, label=f'Mean: {mean_val:.2f}')
ax3.axvline(median_val, color='green', linestyle='--', linewidth=2, label=f'Median: {median_val:.2f}')
ax3.set_xlabel('Values')
ax3.set_ylabel('Data Points')
ax3.set_title('Individual Data Points')
ax3.set_ylim(0.5, 1.5)
ax3.legend()
ax3.grid(True, alpha=0.3)

# 4. Comparison with symmetric distribution
# Create a symmetric version for comparison
symmetric_data = np.random.normal(mean_val, std_val, 1000)
ax4.hist(symmetric_data, bins=30, alpha=0.5, color='gray', label='Normal Distribution', density=True)
ax4.hist(dataset, bins=8, alpha=0.7, color='skyblue', label='Your Data', density=True)
ax4.axvline(mean_val, color='red', linestyle='--', linewidth=2)
ax4.set_xlabel('Values')
ax4.set_ylabel('Density')
ax4.set_title('Your Data vs Normal Distribution')
ax4.legend()
ax4.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# =============================================================================
# EXPLAINING THE POSITIVE SKEW
# =============================================================================
print("\n💡 WHY YOUR DATA IS POSITIVELY SKEWED")
print("-" * 40)

print("🔍 DETAILED EXPLANATION:")
print()

# Calculate mean vs median difference
mean_median_diff = mean_val - median_val
print(f"1️⃣ MEAN vs MEDIAN:")
print(f"   • Mean ({mean_val:.2f}) > Median ({median_val:.2f})")
print(f"   • Difference: {mean_median_diff:.2f}")
print(f"   • This indicates positive skew!")
print()

# Analyze extreme values
print("2️⃣ EXTREME VALUES:")
# Find values that are unusually far from the center
outlier_threshold = mean_val + 1.5 * std_val
potential_outliers = [x for x in dataset if x > outlier_threshold or x < mean_val - 1.5 * std_val]

print(f"   • Looking for values > {outlier_threshold:.1f} or < {mean_val - 1.5 * std_val:.1f}")
if potential_outliers:
    print(f"   • Potential outliers: {potential_outliers}")
else:
    print("   • No clear outliers, but distribution still skewed")

# Analyze the tail
upper_tail = [x for x in dataset if x > q3]
lower_tail = [x for x in dataset if x < q1]

print()
print("3️⃣ TAIL ANALYSIS:")
print(f"   • Upper tail (> Q3): {len(upper_tail)} values: {upper_tail}")
print(f"   • Lower tail (< Q1): {len(lower_tail)} values: {lower_tail}")

# Calculate tail lengths
if upper_tail:
    max_distance_upper = max(upper_tail) - q3
else:
    max_distance_upper = 0

if lower_tail:
    max_distance_lower = q1 - min(lower_tail)
else:
    max_distance_lower = 0

print(f"   • Maximum distance in upper tail: {max_distance_upper}")
print(f"   • Maximum distance in lower tail: {max_distance_lower}")

if max_distance_upper > max_distance_lower:
    print("   • Upper tail is longer → Positive skew!")

# =============================================================================
# WHAT CAUSES THE POSITIVE SKEW
# =============================================================================
print("\n🎯 WHAT CAUSES THE POSITIVE SKEW IN YOUR DATA")
print("-" * 45)

print("Looking at your specific dataset:")
print()

# Frequency analysis
freq_counter = Counter(dataset)
print("📊 VALUE FREQUENCIES:")
for value in sorted(freq_counter.keys()):
    print(f"   • {value}: appears {freq_counter[value]} time(s)")
print()

# Identify the pattern
print("🔍 THE PATTERN:")
print("   • Most values cluster in the 77-88 range")
print("   • A few higher values (95, 97) pull the mean upward")
print("   • The minimum (67) is not as far from the center as the maximum (97)")
print("   • This creates the 'right tail' that defines positive skew")

# Calculate the mathematical reason
print(f"\n🧮 MATHEMATICAL EXPLANATION:")
print("   Skewness formula involves cubing the deviations from the mean:")
print("   • Positive deviations³ = even more positive")
print("   • Negative deviations³ = negative")
print("   • Since you have some large positive deviations (95, 97),")
print("     they dominate when cubed, creating positive skewness")

# Show the actual calculation
deviations = [(x - mean_val) for x in dataset]
cubed_deviations = [(x - mean_val)**3 for x in dataset]

print(f"\n   Example calculations:")
for i, val in enumerate(dataset[:5]):  # Show first 5
    dev = deviations[i]
    cubed = cubed_deviations[i]
    print(f"   • Value {val}: deviation = {dev:.2f}, cubed = {cubed:.2f}")

print(f"\n   Sum of cubed deviations: {sum(cubed_deviations):.2f}")
print(f"   Since this is positive, skewness is positive!")

# =============================================================================
# PRACTICAL IMPLICATIONS
# =============================================================================
print(f"\n🌍 PRACTICAL IMPLICATIONS")
print("-" * 30)

implications = f"""
📋 WHAT THIS MEANS FOR YOUR DATA:

1️⃣ INTERPRETATION:
   • Most values are in the lower-to-middle range (67-88)
   • A few high performers (95, 97) pull the average up
   • Median ({median_val}) better represents "typical" value than mean ({mean_val:.2f})

2️⃣ IF THIS IS TEST SCORES:
   • Most students scored in the 70s-80s
   • A few exceptional students scored in the 90s
   • Average is inflated by high performers

3️⃣ STATISTICAL CONSIDERATIONS:
   • Mean might not be the best measure of central tendency
   • Consider using median for more robust analysis
   • Be aware when applying normal distribution assumptions

4️⃣ DATA VISUALIZATION:
   • Histograms will show longer right tail
   • Box plots will show outliers on the high side
   • Always check for skewness before analysis
"""

print(implications)

# =============================================================================
# SUMMARY
# =============================================================================
print("\n📝 SUMMARY")
print("=" * 20)

summary = f"""
🎯 YOUR DATASET ANALYSIS:

📊 SKEWNESS: {skewness:.4f} (Positive/Right-skewed)

🔍 WHY IT'S POSITIVE:
   • Mean ({mean_val:.2f}) > Median ({median_val})
   • High values (95, 97) create a longer right tail
   • Most data clusters in lower range (67-88)

📏 RANGE: {min_val} to {max_val} (span of {range_val} points)

💡 KEY INSIGHT:
   A few high values pull the distribution to the right,
   creating positive skewness even though most values
   are in the lower portion of the range.
"""

print(summary)

print("\n" + "=" * 60)
print("🎊 Now you understand why your data has positive skewness! 🎊")
print("=" * 60)
````

````python
from scipy.stats import kurtosis
import numpy as np

# Sample data
data = np.array([88, 85, 82, 97, 67, 77, 74, 86, 81, 95, 77, 88, 85, 76, 81])

# Calculate Fisher's kurtosis (excess kurtosis, normal distribution = 0)
fisher_kurtosis = kurtosis(data, fisher=True, bias=False)

# Calculate Pearson's kurtosis (normal distribution = 3)
pearson_kurtosis = kurtosis(data, fisher=False, bias=False)

print(f"Fisher's Kurtosis (Excess Kurtosis): {fisher_kurtosis:.4f}")
print(f"Pearson's Kurtosis: {pearson_kurtosis:.4f}")

# Interpretation based on Fisher's kurtosis
if fisher_kurtosis > 0:
    print("The distribution is leptokurtic (heavier tails than a normal distribution).")
elif fisher_kurtosis < 0:
    print("The distribution is platykurtic (lighter tails than a normal distribution).")
else:
    print("The distribution is mesokurtic (similar tail weight to a normal distribution).")
````
