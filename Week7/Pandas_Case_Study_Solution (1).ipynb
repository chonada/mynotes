```python
import pandas as pd

# Left DataFrame
df1 = pd.DataFrame({
    "emp_id": [1, 2, 3],
    "Name": ["Alice", "Bob", "Charlie"]
}).set_index("emp_id")

# Right DataFrame
df2 = pd.DataFrame({
    "emp_id": [1, 2, 4],
    "Salary": [50000, 60000, 70000]
}).set_index("emp_id")

print("df1:\n", df1, "\n")
print("df2:\n", df2)
```

```python
result = df1.join(df2, how="inner")   # default join is left, here we force inner
print(result)
```

```python
# Merge needs explicit "on" or "left_index/right_index"
result = pd.merge(df1, df2, left_index=True, right_index=True, how="inner")
print(result)
```

```python
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv(r"/content/bank.csv", sep=";")
df.head()
```

```python
# Shape and column types
print(df.shape)
print(df.dtypes)

# Check for missing values
missing = df.isnull().sum()
print("Missing values:\n", missing[missing > 0])

# Check duplicates
print("Duplicate rows:", df.duplicated().sum())
```

```python
# Example: Fill missing categorical with mode
categorical_cols = df.select_dtypes(include='object').columns
for col in categorical_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)

# Detect outliers using IQR for numeric columns
numeric_cols = df.select_dtypes(include=np.number).columns
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers = df[(df[col] < Q1 - 1.5*IQR) | (df[col] > Q3 + 1.5*IQR)]
    print(f"{col} has {outliers.shape[0]} outliers")
```

```python
# Example: Age groups
df['AgeGroup'] = pd.cut(df['age'], bins=[0, 25, 40, 60, 100], labels=['Young','Adult','MidAge','Senior'])

# Example: Balance flag
df['HighBalance'] = df['balance'] > df['balance'].median()

# Interaction feature: duration per contact
df['DurationPerContact'] = df['duration'] / (df['campaign'] + 1)
```

```python
# One-hot encoding for categorical features
df_encoded = pd.get_dummies(df, drop_first=True)
```

```python
# Frequency encoding example
for col in ['job', 'marital']:
    freq = df[col].value_counts() / len(df)
    df[col+'_freq'] = df[col].map(freq)
```

```python
# Average balance by job and marital status
pivot = pd.pivot_table(df, values='balance', index='job', columns='marital', aggfunc=np.mean)
print(pivot)

# Group by multiple columns and aggregate
grouped = df.groupby(['job','education']).agg({
    'balance': ['mean','max','min'],
    'duration': 'mean'
})
print(grouped)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Correlation heatmap for numeric features
plt.figure(figsize=(12,8))
sns.heatmap(df.select_dtypes(include=np.number).corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Numeric Feature Correlations")
plt.show()

# Detect multicollinearity
from statsmodels.stats.outliers_influence import variance_inflation_factor

X = df.select_dtypes(include=np.number).drop(columns=['y'])  # drop target
vif = pd.DataFrame()
vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif["feature"] = X.columns
print(vif.sort_values(by="VIF", ascending=False))
```

```python
# Skewness and kurtosis
skew_kurt = pd.DataFrame()
skew_kurt['skew'] = df.select_dtypes(include=np.number).skew()
skew_kurt['kurtosis'] = df.select_dtypes(include=np.number).kurtosis()
print(skew_kurt)

# Apply log transform for skewed features
for col in ['balance','duration']:
    if abs(df[col].skew()) > 1:
        df[col+'_log'] = np.log1p(df[col])
```

```python
# Distribution of target variable
sns.countplot(x='y', data=df)
plt.title("Target Distribution")
plt.show()

# Target rate by job
job_target = df.groupby('job')['y'].value_counts(normalize=True).unstack()
print(job_target)
```

```python
# Pairplot for selected numeric features
sns.pairplot(df[['age','balance','duration','campaign','y']], hue='y')
plt.show()

# Boxplots by target
sns.boxplot(x='y', y='balance', data=df)
plt.show()

# Heatmap of categorical feature correlations with target
cat_cols = df.select_dtypes(include='object').columns
cat_target_corr = pd.DataFrame({col: df.groupby(col)['y'].apply(lambda x: (x=='yes').mean()) for col in cat_cols})
sns.heatmap(cat_target_corr.T, annot=True, cmap='coolwarm')
plt.show()
```

```python
import pandas as pd
import numpy as np

# Create MultiIndex DataFrame
arrays = [
    ["NY", "NY", "LA", "LA"],
    ["2025-09-25", "2025-09-26", "2025-09-25", "2025-09-26"]
]

index = pd.MultiIndex.from_arrays(arrays, names=("City", "Date"))

df = pd.DataFrame({
    "Sales": [200, 220, 150, 180],
    "Profit": [20, 25, 15, 18]
}, index=index)

print(df)
```

```python
def log_decorator(func):
    def wrapper(*args, **kwargs):
        print(f"Calling function '{func.__name__}' with arguments {args} {kwargs}")
        result = func(*args, **kwargs)
        print(f"Function '{func.__name__}' returned {result}")
        return result
    return wrapper

@log_decorator  # Apply decorator
def add(a, b):
    return a + b

add(5, 3)
```

```python
class CountUp:
    def __init__(self, max_val):
        self.max = max_val
        self.current = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.current > self.max:
            raise StopIteration
        val = self.current
        self.current += 1
        return val

counter = CountUp(5)
for num in counter:
    print(num)
```

```python
def count_up_gen(max_val):
    current = 1
    while current <= max_val:
        yield current
        current += 1

for num in count_up_gen(5):
    print(num)
```

```python
def fibonacci(n):
    a, b = 0, 1
    count = 0
    while count < n:
        yield a
        a, b = b, a + b
        count += 1

# Usage: Print first 10 Fibonacci numbers
for num in fibonacci(10):
    print(num, end=" ")
```
